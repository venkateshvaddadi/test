```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ('RHS-2019-Section-V-Status-of-Health-Manpower-in-Rural-areas-Table-21.csv')
df2=df.drop(df.shape[0]-1)
```

```python
df2
```

```python
required=df2['Required1 - (R)'].to_numpy()
inpos=df2['In Position - (P)'].to_numpy()
state=df2['State/UT']
sno=df2['S. No.']
```

```python
size=required.shape[0]
```

```python
 plt.plot(required, inpos, color='green', linestyle='none',   
         marker='o', markerfacecolor='blue', markersize=5) 
# naming the x axis 
plt.xlabel('Required Positions') 
# naming the y axis 
plt.ylabel('Filled Positions') 
  
# giving a title to my graph 
plt.title('Staus of Health manpower in rural areas') 
```

```python
 start_index=6
 end_index=10
  
# plotting a bar chart 
plt.bar(sno[start_index:end_index],required[start_index:end_index], tick_label = state[start_index:end_index], 
        width = 0.8, color = ['red', 'blue']) 
  
# naming the x-axis 
plt.xlabel('x - axis') 
# naming the y-axis 
plt.ylabel('y - axis') 
# plot title 
plt.title('My bar chart!') 
  
# function to show the plot 
plt.show() 
```

```python
start_index=6
end_index=10
  
plt.bar(sno[start_index:end_index],
        required[start_index:end_index], 
        tick_label = state[start_index:end_index], 
        width = 0.8, 
        color = ['blue'])     
    
# plotting a bar chart 
plt.bar(sno[start_index:end_index],
        inpos[start_index:end_index], 
        tick_label = state[start_index:end_index], 
        width = 0.8, 
        color = ['red']) 

  
# naming the x-axis 
plt.xlabel('State ') 
plt.ylabel('manpower') 
plt.title('Health Manpower in rural areas: Requirements vs Filled') 
plt.legend(['required','in position'] )
plt.show() 
```

```python
index=0
for i in range(start_index,end_index):
    
    plt.bar(index,
        required[i], 
        tick_label = state[i], 
        width = 0.8, 
        color = ['blue'])     
    
    # plotting a bar chart 
    plt.bar(index+1,
        inpos[i], 
        tick_label = state[i], 
        width = 0.8, 
        color = ['red'])
    index=index+2
```
